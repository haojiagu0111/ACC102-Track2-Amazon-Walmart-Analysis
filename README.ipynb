# ACC102 Track 2 Financial Performance Analysis
## Online Retail (Amazon) vs Offline Retail (Walmart)

### Project Overview
This project compares financial differences between global online e-commerce giant Amazon and traditional offline physical retailer Walmart.
It focuses on four key profitability indicators:
- Total Revenue
- Net Profit
- Net Profit Margin
- Year-on-Year Revenue Growth Rate

The research period is from 2015 to 2025, aiming to show how different business models influence corporate revenue and profitability.

### Data Source
All financial data is obtained from **WRDS Compustat Database**, authoritative academic financial statement data of global listed companies.

### Analysis Process
1. Connect WRDS database and extract annual financial data by SQL statements
2. Use Python Pandas to clean data and standardise date format
3. Calculate financial ratios following ACC102 accounting formulas
4. Use Matplotlib to draw annual trend comparison charts
5. Summarise performance gaps between online and offline retail operation models

### Key Findings
- Amazon owns much higher revenue growth speed and obvious profit volatility
- COVID-19 pandemic in 2020 greatly improved Amazon's online business profitability
- Walmart maintains stable low growth and low profit margin, due to high offline store operating costs
- Asset-light online retail model has better growth potential than heavy-asset offline retail

### Project Limitations
- Only overall consolidated financial statements are used, without segmented business income analysis
- Macroeconomic factors, industry policies and exchange rate changes are not included
- More financial efficiency ratios can be added for further comprehensive research

### How to Run
1. Install required Python libraries: wrds, pandas, matplotlib
2. Input personal WRDS account information
3. Run all code cells in Jupyter Notebook in sequence
